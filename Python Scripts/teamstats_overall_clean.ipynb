```python
overallstat = pd.read_csv("statoverall.csv")
overallstat.fillna(0,inplace=True)
overallstat.dtypes
display(overallstat)
```

```python
teamstat = pd.read_csv("teamstats.csv")
display(teamstat)
```

```python
df1=teamstat.drop(teamstat.columns[[8,15]], axis=1) #delete unnamed and NAN columns


df1.drop(['SRS','W','L', 'W/L%','Lg','Finish', 'Playoffs', 'Coaches', 'Top WS'], axis=1, inplace=True)
df1=df1[df1.Season != 'Season']
df1.fillna(0,inplace=True)

df1[['Pace', 'Rel Pace', 'ORtg', 'Rel ORtg', 'DRtg', 'Rel DRtg']]=df1[['Pace', 'Rel Pace', 'ORtg', 'Rel ORtg', 'DRtg', 'Rel DRtg']].apply(pd.to_numeric)
#print (df1[pd.to_numeric(df1.Pace, errors='coerce').isnull()])


print(df1.dtypes)
display(df1)
```

```python
df2 = df1.groupby(['Season'])[['Pace','DRtg','ORtg','Rel Pace','Rel ORtg','Rel DRtg']].mean()
df2
```
