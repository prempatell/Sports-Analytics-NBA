```python
import pandas as pd
```

```python
all_data = pd.read_csv("cumulative_processed_data_22.csv")
all_data.head()
```

```python
#find all players
all_players = all_data['Player'].unique()
```

```python

player_averages = pd.DataFrame(columns=['Age','GS','MP','FG','FGA','FGcent','3P','3PA','3Pcent','2P','2PA','2Pcent','FT','FTA','FTcent','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','GmSc','PointDifference','Player'])
player_std = pd.DataFrame(columns=['Age','GS','MP','FG','FGA','FGcent','3P','3PA','3Pcent','2P','2PA','2Pcent','FT','FTA','FTcent','ORB','DRB','TRB','AST','STL','BLK','TOV','PF','PTS','GmSc','PointDifference','Player'])

i = 0
for player in all_players:
  player_data = all_data[all_data['Player']==player]
  player_averages.loc[i] = player_data.mean(numeric_only=True)
  player_averages.at[i,"Player"] = player

  player_std.loc[i] = player_data.std(numeric_only = True)
  player_std.at[i,"Player"] = player
  i+=1

player_averages["stat"] = "avg"
player_std["stat"] = "std"
```

```python

frames = [player_averages,player_std]
player_stats_combined =pd.concat(frames,)
player_stats_combined.columns=player_stats_combined.columns.str.replace("3","Three")
player_stats_combined.columns=player_stats_combined.columns.str.replace("2","Two")
player_stats_combined.rename(columns={"Player":"PlayerName"},inplace=True)
player_stats_combined.to_csv("player_stats_combined_22.csv",index=False)
#be sure to add "stat" at the top of the columns that avg, str.
player_stats_combined.head(10)
```
